```python
def sistema(ordem):
  """
  sistema do tipo:

  x|y|constantes
  1 -1 = -5
  2  3 = 10



  """


  from numpy import array, linalg, dot
  from fractions import Fraction
  coef_icognita = []
  matriz = []
  matriz_constantes = []
  count = 0
  inv = array([])
  resultado = array([])
  fracao = []

  def racional(numerador: int, denominador: int) ->float:
    return float(numerador / denominador)

# coleta os coeficientes das icognitas e os armazena na matriz deles
  for i in range(1, (ordem**2)+1):
    coef_icognita.append(racional(float(input(f"\nlinha {count+1}: numerador ")), float(input(f"linha {count+1}: denominador "))))
    if (i % ordem ==0):
      count += 1

  for j in range(ordem):
    matriz.append(list(coef_icognita[j*ordem:j*ordem+ordem]))

# coleta as constantes de cada linha e faz a matriz
  count = 0
  for k in range(ordem):
    matriz_constantes.append(racional(float(input(f"\nlinha {count+1}: numerador (constante) ")), float(input(f"\nlinha {count+1}: denominator (constante) "))))
    if (i % ordem ==0):
      count += 1
  matriz = array(matriz)
  matriz_constantes = array(matriz_constantes)
  inv = linalg.inv(matriz)
 # Calcula o resultado das icognitas e transforma em fracao
  resultado = dot(inv, matriz_constantes)
  resultado = resultado.tolist()
  for l in resultado:
    fracao.append(Fraction(l).limit_denominator())
  return fracao

sistema(2)
```
