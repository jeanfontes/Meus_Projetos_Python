```python
#"termo_num" é o numeros de termos da sequencia
#"termo_posicao" coleta o termo que e dado em ordem crescente
#"termos" lista de todos os termos da sequencia
#"polinomio_escrito1" faz os fatores da parte de cima do polinomio expressao
#"polinomio_escrito2" e o 01 sem o asterisco final
#"fator_cima" separa os fatores da parte de cima do pilinomio expressao,
#"polinomio_parcial_cima_list"  contem o "numerador" (sem coeficiente lider) de cada polinomio que gera cada termo
#"trocar" variavel auxiliar para trocar "x" pelo numero que n zera o poliniomui, a fim de dar 1
#"polinomio_parcial_baixo_list" contem o denominador de cada polinomio que gera cada termo
#"sobre" variavel responsavel por fazer quociente de cada termo dos polinomios parciais com seu respectivo parceiro
#"polinomio_parcial_pronto" contem o quociente anterior (de cada termo com seu respectivo parceiro) em formato de str
#"asterisco" variavel responsavel por colocar o sinal de multiplicacao no polinomio parcial
#"polinomio_parcial_pronto2" contem o quociente anterior (de cada termo com seu respectivo parceiro) em formato de str e com sinal de multiplicacao (*)
#"expr" varaval responsavel por transformar cada pedaco do polinomio_parcial_pronto2 em expressao sympy
#"polinomio_expr" tem cada pedaco do polinomio_parcial_pronto2 em expressao sympy
#"expr2" tem cada pedaco de polinomio_expr multiplicado por seu termo de "termos"
#"polinomio_expr2" tem todos "expr2", pode-se dizer que e o polinomio finalizado (seus pedacos)
#"soma" e o polinomio (sequencia) realmente finalizados (e a soma das partes de polinomio_expr2)
#"list_frac" lista de todos os coefientes da sequencia na forma decimal
#"list_frac_pronto" e o coeficientes anteriores da forma de fracao
import sympy as sp
from sympy.polys.polytools import poly
from fractions import Fraction
from IPython.display import clear_output
termos=[]
polinomio_parcial_cima_list=[]
polinomio_parcial_baixo_list=[]
fator_cima=[]
polinomio_parcial_pronto=[]
polinomio_parcial_pronto2=[]
polinomio_expr=[]
polinomio_expr2=[]
x=sp.symbols("x")
polinomio_escrito1="(x-s)*"
polinomio_escrito2=""
polinomio_escrito3=""
juntar=juntar2=trocar=sobre=asterisco=expr=expr2=juntar_poly_frac=""
soma=0
list_frac=[]
list_frac_pronto=[]
# coletando os termos da sequencia
termos_num=int(input("Digite quantos termos terá a sequencia"))
for i in range(termos_num):
  print("Digite o termo",i+1)
  termo_posicao=float(input())
  termos.append(termo_posicao)
for j in range(1,termos_num+1):
  polinomio_escrito2+=polinomio_escrito1
  polinomio_escrito2=polinomio_escrito2.replace("s",str(j))
polinomio_escrito2=polinomio_escrito2[:len(polinomio_escrito2)-1]
fator_cima=polinomio_escrito2.split("*")
#fazendo parte de cima dos polinomios
for k in fator_cima:
  juntar+=k
for l in fator_cima:
  polinomio_parcial_cima_list.append(juntar.replace(str(l),""))
#fazendo parte de baixo dos polinomios
for m in range(len(polinomio_parcial_cima_list)):
  trocar=""
  trocar=polinomio_parcial_cima_list[m]
  trocar=trocar.replace("x",(str(m+1)))
  polinomio_parcial_baixo_list.append(trocar)
#polinomio cima sobre baixo
for n in range(len(polinomio_parcial_cima_list)):
  sobre=""
  sobre=sobre.join(("(",polinomio_parcial_cima_list[n],")/(",polinomio_parcial_baixo_list[n],")"))
  polinomio_parcial_pronto.append(sobre)
for o in polinomio_parcial_pronto:
  asterisco=o
  asterisco=asterisco.replace(")(",")*(")
  polinomio_parcial_pronto2.append(asterisco)
#polinomio expressao
for p in polinomio_parcial_pronto2:
  expr=p
  expr=sp.parsing.sympy_parser.parse_expr(expr)
  polinomio_expr.append(expr)
for q in range(len(polinomio_expr)):
  expr2=""
  expr2=polinomio_expr[q]*termos[q]
  polinomio_expr2.append(expr2)
for r in polinomio_expr2:
  soma+=r
soma=sp.simplify(soma)
clear_output()
#coletar coeficientes
soma=poly(soma)
list_frac=soma.all_coeffs()
#decimal pra fracao
for s in list_frac:
  list_frac_pronto.append(str(Fraction(str(s)).limit_denominator()))
#polinomio fracao para expressao
print("Grau:",len(list_frac_pronto)-1)
for t in range(len(list_frac_pronto)-1,-1,-1):
  print("\n"+"("+str(list_frac_pronto[len(list_frac_pronto)-1-t])+")","expoente",t)

```
