```python
from datetime import date, timedelta
def angel(data: str) -> int:
  """

  calculo o numero de angel de uma data

  data é no formato brasileiro (dd/mm/aaaa)
  o retorno é o número de angel
  soma_digitos é a soma de digitos da data na iteraçao 1
  soma_digitos2 é a soma dos digitos de maneira intermediaria
  soma_digitos_pronto é a soma dos digitos que é testada para o objetivo

  """
  data = data.replace("/","")# deixa só os digitos da data
  # variaveis do processamente \/
  soma_digitos = soma_digitos2 = soma_digitos_pronto = 0
  while (len(str(soma_digitos_pronto)) != 1) or (soma_digitos == 0):
    # faz a primeira soma dos digitos \/
    if (soma_digitos == 0):
      for i in data:
        soma_digitos += int(i)
      soma_digitos_pronto = soma_digitos
    # faz somas intermediarias ate atinjir objetivo \/
    for i in str(soma_digitos_pronto):
      soma_digitos2 += int(i)
    soma_digitos_pronto = soma_digitos2
    soma_digitos2 = 0
  return int(str(soma_digitos_pronto)*3)


def data_num(data:str)-> int:
  """
  converte data em numero (diferença de dias de 01/01/1900)
  data é a data a ser convertida
  data_partida a mesma coisa que variavel data, mas pro processamento
  data_base data de referencia (01/01/1900)

  """
  data_base = date(1900, 1, 1)
  #entrada convertida para uma data valida da datetime
  data_partida = date(*[int(i) for i in data.split("/")[::-1]])
  return (data_partida - data_base).days

def num_data(num: int)->str:
  """
  transforma o numero de data numa data
  num é a data formatada como número
  data_base é a data de referencia a data tal que possui o dias discorridos
  data_retorno é o numero convertido em data


  """

  data_base = date(1900, 1, 1)
  # data base + dias corridos da base
  data_retorno = data_base + timedelta(days=num)
  return f"{data_retorno.day}/{data_retorno.month}/{data_retorno.year}"




def encontrar(anjo:int, data_chute: str)->str:
  """
  encontra uma data que dê o número de anjo fornecido
  argumentos:
  anjo- numero a encontrar para uma data

  data_chute- ponto de partida


  """


  x0 = data_num(data_chute) # chute para o metodo de newton
  data_encontrada = "" # resposta da pergunta
  def num_angel(num:int)->int:
    """
    calcula o numero de anjo de uma data convertida em numero.
    função para o metodo de newton

    """
    return angel(num_data(num)) - anjo
  def newton(func, chute: int)->int:
    """
    aplica o metodo de newton para num_angel
    considera uma função generica a principio

    argumentos:
    func função que se aplica o metodo de newton
    chute ponto inicial no caso uma data formatada em numero

    variaveis processamento:
    xn compotente da formula do metodo de newton
    data_encontrada é o zero da função. no caso a data que resulta no numero de angel

    """
    xn = chute
    def dydx(f:"funcao",x: "ponto")->int:
      """
      calcula a derivada em um ponto x

      argumentos:
      f: funcao a derivar
      x: ponto


      """
      return (f(x+1)-f(x))/1
    for i in range(100):
      xn = xn - (func(xn))/(dydx(func,xn))
    return xn

  data_encontrada = num_data(newton(num_angel,x0))
  return data_encontrada,"Encontrado: sim"if (angel(data_encontrada) - anjo == 0) else "Encontrado: não"


encontrar(444,"22/04/1998" )
```
